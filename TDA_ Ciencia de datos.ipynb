```python
# 1 importar pandas
import pandas as pd

```

```python
# 2. Importar JSON desde la URL
url= "https://www.datos.gov.co/resource/uqkt-7tia.json"
df_json = pd.read_json(url)
print(df_json)
```

```python
print(df_json.head())
```

```python
# 4. Ver las últimas filas

print(df_json.tail())  #muestra las últimas 5 filas
```

```python
# 5. Información del dataframe (tipos de datos, nulos, etc.)

print(df_json.info())
```

```python
# 6. Estadística básica

print(df_json.describe(include='all'))
```

```python
# 7. Ver nombres de columnas

print(df_json.columns)
```

```python
# 8. Ver índices

print(df_json.index)
```

```python
# 9. Seleccionar una columna

print(df_json["producto"])
```

```python
# 10. Seleccionar varias columnas

print(df_json[["producto", "marca", "estado_registro"]])
```

```python
# 11. Seleccionar una fila (por posición)

print(df_json.iloc[0])
```

```python
# 12. Seleccionar filas con columnas

print(df_json.iloc[0:5, 0:4])
```

```python
# 13. Filtrar datos (registros VIGENTES)

vigentes = df_json[df_json["estado_registro"] == "VIGENTE"]
print(vigentes)
```

```python
# 14. Agregar una nueva columna (longitud del nombre del producto)

df_json["longitud_producto"] = df_json["producto"].apply(len)
print(df_json[["producto", "longitud_producto"]])
```

```python
# 15. Modificar valores (convertir ciudad_titular a minúsculas)

df_json["ciudad_titular"] = df_json["ciudad_titular"].str.title()
print(df_json["ciudad_titular"])
```

```python
# 16. Eliminar columna (longitud_producto)

df_json = df_json.drop("longitud_producto", axis=1)
print(df_json.columns)
```

```python
# 17. Contar valores nulos por columna

print(df_json.isnull().sum())
```

```python
# 18. Llenar valores nulos en columnas de texto con 'SIN DATO'

df_json["marca"] = df_json["marca"].fillna("SIN DATO")
df_json["forma_farmaceutica"] = df_json["forma_farmaceutica"].fillna("SIN DATO")
print(df_json[["marca", "forma_farmaceutica"]].head(10))
```

```python
# 19. Ordenar datos por fecha de vencimiento (ascendente)

order = df_json.sort_values("fecha_vencimiento")
print(order[["producto", "marca", "fecha_vencimiento"]].head(10))
```

```python
# 20. Ordenar descendente (los que vencen más tarde primero)

order_desc = df_json.sort_values("fecha_vencimiento", ascending=False)
print(order_desc[["producto", "marca", "fecha_vencimiento"]].head(10))
```

```python
# 21. Agrupar datos: contar cuántos productos hay por modalidad

group = df_json.groupby("modalidad")["producto"].count()
print(group)
```

```python
# 22. Contar registros: cuántos productos hay por estado de registro

count = df_json["estado_registro"].value_counts()
print(count)
```
